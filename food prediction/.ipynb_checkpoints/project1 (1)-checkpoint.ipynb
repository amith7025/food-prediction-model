```python
import torchvision
from torchvision import transforms
```

```python
train_transforms = transforms.Compose([
    transforms.Resize(size=(64,64)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
    transforms.RandomRotation(15),  # Rotate images by up to 15 degrees
    transforms.RandomResizedCrop(64, scale=(0.8, 1.2), ratio=(0.8, 1.2)),
    transforms.ToTensor(),
    #transforms.Normalize(mean=[0.485, 0.456, 0.406],  # Mean values for each channel
                         #std=[0.229, 0.224, 0.225])
])
```

```python
test_transforms = transforms.Compose([
    transforms.Resize(size=(64,64)),
    transforms.ToTensor(),
    #transforms.Normalize(mean=[0.485, 0.456, 0.406],  # Mean values for each channel
                         #std=[0.229, 0.224, 0.225])
])
```

```python
from torchvision import datasets
from torch.utils.data import DataLoader
```

```python
train_data = datasets.ImageFolder(
    root=train_dir,
    transform=train_transforms,
    target_transform=None
)
```

```python
test_data = datasets.ImageFolder(
    root=test_dir,
    transform=test_transforms,
    target_transform=None
)
```

```python
BATCH_SIZE = 8
train_loader = DataLoader(train_data, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_data,batch_size=BATCH_SIZE,shuffle=False)
```

```python
import torch.nn as nn
```

```python
from torch.nn.modules.conv import Conv2d
class FoodClassification(nn.Module):
  def __init__(self,input_shape=3,output_shape=len(classes)):
    super().__init__()

    self.Conv_blk_1 = nn.Sequential(
        nn.Conv2d(3,64,kernel_size=2,stride=1),
        nn.ReLU(),
        #nn.Dropout2d(0.5),
        nn.MaxPool2d(kernel_size=2,stride=1),
        nn.Conv2d(64,32,kernel_size=2,stride=1),
        nn.ReLU(),
        nn.MaxPool2d(kernel_size=2,stride=2)
    )
    self.avgpool = nn.AvgPool2d(kernel_size=2)
    self.classifier = nn.Sequential(
        nn.Dropout(),
        nn.Linear(32*15*15,32),
        nn.ReLU(),
        nn.Dropout(),
        nn.Linear(32,32),
        nn.ReLU(),
        nn.Linear(32,output_shape)
    )
  def forward(self, x):
        x = self.Conv_blk_1(x)
        x = self.avgpool(x)
        x = torch.flatten(x, 1)
        x = self.classifier(x)
        return x
```

```python
torch.manual_seed(42)

model = FoodClassification().to(device)
```

```python
image = torch.randn(8,3,64,64)
model(image.to(device))
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(params=model.parameters(),lr=0.001,weight_decay=0.0001)
```

```python
def train_step(model,dataloader,loss_fn,optimizer,device):
  train_loss = 0
  model.train()
  for batch in dataloader:
    optimizer.zero_grad()
    input,target = batch
    input,target = input.to(device),target.to(device)
    output = model(input)
    loss = loss_fn(output,target)
    loss.backward()
    optimizer.step()
    train_loss += loss.data.item()
  train_loss /= len(dataloader)
  return train_loss
```

```python
def test_step(model,dataloader,loss_fn,device):
  test_loss = 0
  model.eval()
  for batch in dataloader:
    input,target = batch
    input,target = input.to(device),target.to(device)
    output = model(input)
    loss = loss_fn(output,target)
    test_loss += loss.data.item()
  test_loss /= len(dataloader)
  return test_loss
```

```python
from tqdm.auto import tqdm
```

```python
epochs = 50
```

```python
for epoch in tqdm(range(epochs)):
  train_loss = train_step(
      model=model,
      dataloader=train_loader,
      loss_fn=loss_fn,
      optimizer=optimizer,
      device=device
  )
  test_loss = test_step(
      model=model,
      dataloader=test_loader,
      loss_fn=loss_fn,
      device=device
  )
  print(f"train loss:{train_loss}  test loss:{test_loss}")
  if test_loss <= 0.6:
    break
```
